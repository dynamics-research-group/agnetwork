```python
import time
import json

timestamp_conversion_factor = 10**9

json_import_directory = "/Users/Julian/Documents/WorkDocuments/Irreducible Element/IE models/json/"
json_export_directory = "/Users/Julian/Documents/WorkDocuments/Irreducible Element/IE models/valid-json/"

json_files_to_convert = [{"name" : "Aeroplane 1.json", 
						  "type": "free"}]

population = "test"

for json_file in json_files_to_convert:
	# print(json_file)
	file_path = f"{json_import_directory}{json_file['name']}"
	with open(file_path, "r") as infile:
		old_json = json.load(infile)

	valid_json = {}

	timestamp = time.time() * timestamp_conversion_factor
	valid_json["timestamp"] = timestamp

	valid_json["population"] = population

	try:
		valid_json["name"] = old_json["name"]
	except NameError:
		print("name not defined for the chosen structure")

	if "irreducible_element_model" in old_json:
		valid_json["models"] = {"irreducibleElement": 
									{"type": json_file['type']}
								}
		
		valid_json["elements"] = []
		for old_element in old_json["irreducible_element_model"]["elements"]:
			new_element = {}
			new_element["name"] = old_element["name"]
			# Convert old type descriptions to updated schema types
			if old_element["type"] == 'regular':
				new_element["type"] = "regular"
			elif old_element["type"] == 'boundary-condition':
				new_element["type"] = "ground"
			new_element["contextual"] = {"type": old_element["description"]}
			new_element["coordinates"] = {"global": {
												"translational": {
													"x": {
														"unit": "mm",
														"value": 0
													},
													"y": {
														"unit": "mm",
														"value": 0
													},
													"z": {
														"unit": "mm",
														"value": 0
													}
												},
												"rotational": {
													"alpha": {
														"unit": "degrees",
														"value": 0
													},
													"beta": {
														"unit": "degrees",
														"value": 0
													},
													"gamma": {
														"unit": "degrees",
														"value": 0
													}
												}
                        					}}

			valid_shape_classes = ["beam","plate","shell","solid"]
			try:
				if old_element["shape"]["class"] in valid_shape_classes:
					new_element["geometry"] = {"type": {
													"name": old_element["shape"]["class"],
												}}
			except KeyError:
				print(f"No shape information for element {old_element['name']}")

				


			valid_json["elements"].append(new_element)

			


	with open (f"{json_export_directory}/valid-{json_file['name']}", "w") as outfile:
		json.dump(valid_json, outfile, indent=4)
	

```
